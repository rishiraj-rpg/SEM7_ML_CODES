```python
import matplotlib.pyplot as mp
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("exp6.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
inputs = df.drop('Survived',axis='columns')
inputs
```

```python
target = df.Survived
target
```

```python
#one hot encoding
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer 

final_input = pd.get_dummies(inputs, columns = ['Sex'])
```

```python
final_input
```

```python
final_input.Age[:10]
```

```python
#filling the missing values
final_input.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
final_input.head()
```

```python
# normalization

from sklearn import preprocessing
d = preprocessing.normalize(final_input)
final_df = pd.DataFrame(d, columns=["Pclass","Age","Fare","Sex_female","Sex_male"])
final_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_df,target,test_size=0.2)
print(X_train)
print(y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from vecstack import stacking
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.feature_selection import RFE

from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
```

```python
model_1 = LogisticRegression()
model_2 = KNeighborsClassifier()
model_3 = DecisionTreeClassifier()
```

```python
y_train = np.array(y_train)
```

```python
from math import sqrt
#Stacking
all_models = [model_1, model_2, model_3]

s_train, s_test = stacking(all_models, X_train,y_train,X_test,regression=True,random_state=None)
 
final_model = model_1

final_model = final_model.fit(s_train, y_train)
 
pred_final = final_model.predict(s_test)
 
print("ROOT MEAN SQUARED ERROR =",sqrt(mean_squared_error(y_test, pred_final)))

print("ACCURACY =",accuracy_score(y_test,pred_final))
```

```python
#Bagging
clf = RandomForestClassifier(n_estimators = 50) 
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
 
print("ROOT MEAN SQUARED ERROR =",sqrt(mean_squared_error(y_test, y_pred)))
 
print("ACCURACY =", accuracy_score(y_test, y_pred))
```

```python
#Boosting
xgb_classifier = xgb.XGBClassifier()
xgb_classifier.fit(X_train,y_train)

y_pred = xgb_classifier.predict(X_test)

print("ROOT MEAN SQUARED ERROR =",sqrt(mean_squared_error(y_test, y_pred)))
print("ACCURACY =", accuracy_score(y_test, y_pred))
```
